```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Peripheral Point

```python
np.random.seed(42)
num_points = 100
time = np.arange(0, num_points)
trend_slope = 0.1
seasonal_period = 12
noise_scale = 2.5
trend = trend_slope * time
seasonality = 10 * np.sin(2 * np.pi * time / seasonal_period)
noise = np.random.normal(loc=0, scale=noise_scale, size=num_points)
time_series = trend + seasonality + noise
anomaly_index = np.random.randint(40, 60)
anomaly_magnitude = 30
time_series[anomaly_index] += anomaly_magnitude
df = pd.DataFrame({'Time': time, 'Value': time_series})
plt.figure(figsize=(12, 6))
plt.plot(df['Time'], df['Value'], label='Time Series')
plt.scatter(df.loc[anomaly_index, 'Time'], df.loc[anomaly_index, 'Value'], color='red', marker='x', s=100, label='Anomaly')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Data with a Peripheral Point Anomaly')
plt.legend()
plt.grid(True)
plt.show()
```

## Local Density anomaly

```python
num_points = 100
time = np.arange(0, num_points)
trend_slope = 0.1
seasonal_period = 12
noise_scale = 2.5
trend = trend_slope * time
seasonality = 10 * np.sin(2 * np.pi * time / seasonal_period)
noise = np.random.normal(loc=0, scale=noise_scale, size=num_points)
time_series = trend + seasonality + noise
anomaly_start = 60
anomaly_end = 70
anomaly_magnitude = 25
time_series[anomaly_start:anomaly_end] += anomaly_magnitude
df = pd.DataFrame({'Time': time, 'Value': time_series})
plt.figure(figsize=(12, 6))
plt.plot(df['Time'], df['Value'], label='Time Series')
plt.fill_between(df['Time'].iloc[anomaly_start:anomaly_end], df['Value'].iloc[anomaly_start:anomaly_end],
                 color='red', alpha=0.3, label='Local Density Anomaly')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Data with Trend, Seasonality, Noise, and a Local Density Anomaly')
plt.legend()
plt.grid(True)
plt.show()
```

### A global Density anoamly

```python
np.random.seed(42)
num_points = 100
time = np.arange(0, num_points)
trend_slope = 0.1
seasonal_period = 12
noise_scale = 2.5

trend = trend_slope * time

seasonality = 10 * np.sin(2 * np.pi * time / seasonal_period)
noise = np.random.normal(loc=0, scale=noise_scale, size=num_points)

time_series = trend + seasonality + noise
anomaly_magnitude = 30
time_series_with_anomaly=time_series+anomaly_magnitude
df = pd.DataFrame({'Time': time, 'Value': time_series})
plt.figure(figsize=(12, 6))
plt.plot(df['Time'], df['Value'],label="Normal Time series data",color="blue")
plt.plot(df['Time'], time_series_with_anomaly, label='Time Series with Global Density Anomaly',color="orange")
plt.axhline(y=anomaly_magnitude, color='red', linestyle='dashed', label='Global Density Anomaly')

plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Global Density Anomaly')
plt.legend()
plt.grid(True)
plt.show()
```

## A local Additive anomaly

```python
np.random.seed(42)
num_points = 100
time = np.linspace(0, 1, num_points)
time_series = np.sin(2*np.pi * time)

anomaly_index = np.where(time_series==np.max(time_series))
print(anomaly_index)
time_series[anomaly_index] -= 0.3

df = pd.DataFrame({'Time': time, 'Value': time_series})

plt.figure(figsize=(12, 6))
plt.plot(df['Time'], df['Value'], label='Time Series with Local Additive Anomaly', color="blue")
plt.scatter(time[anomaly_index], time_series[anomaly_index], color="red", label='Local Additive Anomaly',alpha=1)

plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Data with a Local Additive Anomaly')
plt.legend()
plt.grid(True)
plt.show()
```

## Deviant cycle

```python
num_points = 100
time = np.arange(0, num_points)
trend_slope = 0.1
seasonal_period = 12
noise_scale = 2.5
trend = trend_slope * time
seasonality = 10 * np.sin(2 * np.pi * time / seasonal_period)
noise = np.random.normal(loc=0, scale=noise_scale, size=num_points)
time_series = trend + seasonality + noise
anomaly_index_start = 40
anomaly_index_end=60
noise_scale_anomaly=0.5
size_anomaly=anomaly_index_end-anomaly_index_start
time_series[anomaly_index_start:anomaly_index_end] =np.random.normal(loc=0, scale=noise_scale_anomaly, size=size_anomaly)
plt.figure(figsize=(12, 6))
plt.plot(time, time_series, label='Time Series')
plt.scatter(time[anomaly_index_start:anomaly_index_end], time_series[anomaly_index_start:anomaly_index_end], color='red', marker='*', s=100, label='Deviant cycle')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Data with a Deviant Cycle')
plt.legend()
plt.grid(True)
plt.show()
```

## Temporary Change

```python
num_points = 100
time = np.linspace(0,1,num_points)
time_series = np.sin(2 * np.pi * time)
noise = np.random.normal(loc=0, scale=0.1, size=num_points)
time_series = time_series+ noise
anomaly_index_start = 40
anomaly_index_end=60
noise_anomaly_scale=0.5
noise_anomaly_size=anomaly_index_end-anomaly_index_start
time_series[anomaly_index_start:anomaly_index_end] +=5+np.random.normal(loc=0, scale=noise_anomaly_scale, size=noise_anomaly_size)
plt.figure(figsize=(12, 6))
plt.plot(time, time_series, label='Normal Time Series Data', color='blue')
plt.fill_between(time[anomaly_index_start:anomaly_index_end], time_series[anomaly_index_start:anomaly_index_end], color='red', label='Temporary Change Anomaly',alpha=0.5)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Data with a Temporary Change Anomaly')
plt.legend()
plt.grid(True)
plt.show()
```

## Level Shift

```python
num_points = 100
time = np.linspace(0,1,num_points)
noise = np.random.normal(loc=0, scale=0.1, size=num_points)
time_series = time+noise
time_series = time_series
anomaly_index_start = 40
anomaly_index_end=80
noise_anomaly_scale=0.1
noise_anomaly_size=anomaly_index_end-anomaly_index_start
time_series[anomaly_index_start:anomaly_index_end] =np.power(1.5*time[anomaly_index_start:anomaly_index_end],5)
size_anomaly=len(time_series)-anomaly_index_end
time_series[anomaly_index_end:] =3+np.random.normal(loc=0, scale=noise_anomaly_scale, size=size_anomaly)
plt.figure(figsize=(12, 6))
plt.plot(time, time_series, label='Normal Time Series Data', color='blue')
plt.scatter(time[anomaly_index_start:anomaly_index_end], time_series[anomaly_index_start:anomaly_index_end], color='red', label='Level shift',alpha=1)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Data with a Level shift Anomaly')
plt.legend()
plt.grid(True)
plt.show()
```

## Innovational outlier

```python
num_points = 100
time = np.linspace(0,1,num_points)
time_series = np.sin(10* np.pi * time)
noise = np.random.normal(loc=0, scale=0.1, size=num_points)
time_series = time_series+ noise
anomaly_index_start = 40
anomaly_size=len(time_series)-anomaly_index_start
anomaly_noise=np.random.normal(loc=0, scale=0.6, size=anomaly_size)
time_series[anomaly_index_start:]=6*np.sin(20*np.pi*time[anomaly_index_start:])+anomaly_noise
plt.figure(figsize=(12, 6))
plt.plot(time[:anomaly_index_start], time_series[:anomaly_index_start], label='Normal Time Series Data', color='blue')
plt.plot(time[anomaly_index_start:], time_series[anomaly_index_start:], color='red', label='Innovational Outlier',alpha=0.5)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Data with an Innovational outlier Anomaly')
plt.legend()
plt.grid(True)
plt.show()
```

## Trend change

```python
n_samples = 200
change_point = n_samples // 2
slope1 = 0.3
intercept1 = 2.0
trend1 = np.linspace(0, change_point - 1, change_point) * slope1 + intercept1
slope2 = -0.1
intercept2 = 8.0
trend2 = np.linspace(change_point, n_samples - 1, n_samples - change_point) * slope2 + intercept2
time_series_data = np.concatenate((trend1, trend2))
noise_level = 1.0
time_series_data += np.random.randn(n_samples) * noise_level
plt.figure(figsize=(10, 5))
plt.plot(range(n_samples), time_series_data)
plt.axvline(change_point, color='red', linestyle='--', linewidth=2, label='Trend Change Point')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.title('Time Series Data with Trend Change')
plt.show()
```

## Variation Change

```python
n_samples = 200
change_point = n_samples // 2
baseline_mean = 10.0
baseline_std = 1.0
baseline_data = np.random.normal(loc=baseline_mean, scale=baseline_std, size=n_samples)
anomaly_mean = 10.0
anomaly_std = 5.0
anomaly_data = np.random.normal(loc=anomaly_mean, scale=anomaly_std, size=n_samples - change_point)
time_series_data = np.concatenate((baseline_data[:change_point], anomaly_data))
noise_level = 1.0
time_series_data += np.random.randn(n_samples) * noise_level
plt.figure(figsize=(10, 5))
plt.plot(range(n_samples), time_series_data)
plt.axvline(change_point, color='red', linestyle='--', linewidth=2, label='Variation Change Point')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.title('Time Series Data with Variation Change Anomaly')
plt.show()
```
